```python
import pandas as pd
import csv
print("Step 1: Loading the dataset...")
df = pd.read_csv(
    r"olist_order_reviews_dataset.csv",
    encoding="utf-8",
    low_memory=False
)

# Store original row count
original_row_count = len(df)
print(f"Original row count: {original_row_count}")

# Step 2: Select only the required columns
print("\nStep 2: Selecting required columns...")
required_columns = ['review_id', 'order_id', 'review_score', 'review_creation_date', 'review_answer_timestamp']
df = df[required_columns]

print(f"Columns retained: {list(df.columns)}")

# Step 3: Handle missing values by filling with empty strings or appropriate defaults
print("\nStep 3: Handling missing values...")
df = df.fillna('')

# Step 4: Count duplicates before removing
print("\nStep 4: Removing duplicates...")
duplicates_count = df.duplicated().sum()
print(f"Number of duplicate records found: {duplicates_count}")

# Remove duplicate rows, keeping the first occurrence
df = df.drop_duplicates(keep='first')

# Step 5: Final row count after cleaning
final_row_count = len(df)
print(f"Final row count after cleaning: {final_row_count}")
print(f"Rows removed: {original_row_count - final_row_count}")

# Step 6: Display summary information
print("\n" + "="*60)
print("CLEANING SUMMARY")
print("="*60)
print(f"Original rows: {original_row_count}")
print(f"Final rows: {final_row_count}")
print(f"Duplicates removed: {duplicates_count}")
print(f"Column names: {list(df.columns)}")
print(f"Data types:\n{df.dtypes}")
print(f"\nFirst few rows:\n{df.head()}")

# Step 7: Save the cleaned dataset as CSV with UTF-8 encoding
print("\n" + "="*60)
print("Step 7: Saving cleaned dataset...")
output_file = 'olist_order_reviews_clean.csv'
df.to_csv(
    output_file,
    index=False,
    encoding='utf-8',
    quoting=csv.QUOTE_MINIMAL
)
print(f" Cleaned dataset saved as: {output_file}")
print("="*60)
```
